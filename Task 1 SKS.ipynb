# Task 1 : Data Cleaning and Preprocessing

## Step 1 : Import Libraries

```python
# Data Manipulation
import pandas as pd
import numpy as np

# Data Visualization (used in later tasks)
import matplotlib.pyplot as plt
import seaborn as sns

# Display Settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
```

#### pd.set_option() → Ensures all columns are visible while working.

## Step 2 : Load Dataset

```python
# Load Dataset
df = pd.read_csv("Telco_Customer_Churn_Dataset .csv")

# Display first 5 rows
df.head()
```

## Step 3 : Basic Information

```python
print("Dataset Shape:", df.shape)

df.info()
```

```python
df.describe()
```

## Step 4 : Check Duplicate Record

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows :", duplicates)
```

## Step 5 : Check Missing Values

```python
df.isnull().sum()
```

```python
# Replace blank spaces with NaN
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)
```

## Step 6: Convert TotalCharges into Numeric

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

#### This will generate NaN values for blank entries.

## Step 7 : Check Missing Values Again

```python
df.isnull().sum()
```

## Step 8 : Handle Missing Values

```python
df.dropna(inplace=True)

print(df.shape)
```

## Step 9 : Verify Data Types

```python
df.info()
```

## Step 10 : Explore Categorical Columns

```python
categorical_cols = df.select_dtypes(include="object").columns

print(categorical_cols)
```

## Step 11 : Drop customerID

#### Customer ID is unique for every customer and doesn't help in predicting churn.

```python
df.drop("customerID", axis=1, inplace=True)
```

```python
df.columns
```

## Step 12 : Encode Target Variable

#### Convert the target variable (Churn) to numeric.

```python
df["Churn"] = df["Churn"].map({
    "No": 0,
    "Yes": 1
})
```

```python
print(df['Churn'])
```

## Step 13 : One-Hot Encode Remaining Categorical Variables

```python
df = pd.get_dummies(
    df,
    drop_first=True,
    dtype=int
)
```

#### Using drop_first=True avoids redundant dummy variables and reduces multicollinearity in many machine learning models.
