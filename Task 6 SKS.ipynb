# Task 6 : Advanced Visualizations

## Step 1 : Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
sns.set_theme(style='whitegrid')
```

## Step 2 : Load Dataset

```python
df = pd.read_csv("Telco_Customer_Churn_Dataset .csv")
df['TotalCharges'] = df['TotalCharges'].replace(' ',np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

df.dropna(inplace=True)
```

## Visualization 1 : Box Plot
#### Monthly Charges vs Churn

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Churn",
    hue='Churn',
    y="MonthlyCharges",
    palette="Set2"
)

plt.title("Monthly Charges vs Churn")

plt.show()
```

### Insight

- Customers who churn generally have higher monthly charges than customers who stay.

## Visualization 2 : Violin Plot

```python
plt.figure(figsize=(8,5))

sns.violinplot(
    data=df,
    x="Churn",
    hue='Churn',
    y="tenure",
    palette="Pastel1"
)

plt.title("Tenure Distribution by Churn")

plt.show()
```

### Insight

- Churned customers are concentrated in the early months of their subscription.

## Visualization 3 : Pair Plot

```python
sns.pairplot(
    df,
    vars=[
        "tenure",
        "MonthlyCharges",
        "TotalCharges"
    ],
    hue="Churn"
)

plt.show()
```

### Insight

- The pair plot shows how the three numerical variables relate to each other while highlighting churn patterns.

## Visualization 4 : Correlation Heatmap

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=["int64","float64"])

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Insight

- Tenure and Total Charges have a strong positive correlation.

## Visualization 5 : Contract Type vs Monthly Charges

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Contract",
    hue='Contract',
    y="MonthlyCharges",
    palette="Set3"
)

plt.title("Monthly Charges Across Contract Types")

plt.xticks(rotation=15)

plt.show()
```

### Insight

- Monthly charges vary across contract types, with month-to-month customers showing greater variability.

## Visualization 6: Internet Service Distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="InternetService",
    hue="Churn",
    palette="Dark2"
)

plt.title("Internet Service vs Churn")

plt.show()
```

### Insight

- Fiber Optic customers exhibit a higher churn count than DSL users.

## Visualization 7 : Monthly Charges Histogram

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="MonthlyCharges",
    bins=30,
    kde=True,
    color="green"
)

plt.title("Distribution of Monthly Charges")

plt.show()
```

### Insight

- Monthly charges are spread across a wide range, with many customers paying moderate to high monthly fees.

## Visualization 8 : Total Charges Histogram

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="TotalCharges",
    bins=30,
    kde=True,
    color="orange"
)

plt.title("Distribution of Total Charges")

plt.show()
```

### Insight

- Total charges are right-skewed because long-term customers accumulate higher charges over time.

## Visualization 9 : Tenure Histogram

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="tenure",
    bins=30,
    kde=True,
    color="steelblue"
)

plt.title("Customer Tenure Distribution")

plt.show()
```

### Insight

- Many customers either leave early or remain with the company for a long period.

## Visualization 10 : Churn Percentage by Contract

```python
contract = pd.crosstab(
    df["Contract"],
    df["Churn"],
    normalize="index"
) * 100

contract.plot(
    kind="bar",
    stacked=True,
    figsize=(8,5)
)

plt.title("Churn Percentage by Contract")

plt.ylabel("Percentage")

plt.show()
```

### Insight

- Month-to-month customers have the highest percentage of churn, reinforcing the findings from Task 2.

## Advanced Visualization - Conclusion

- The visualizations provided deeper insights into customer behavior and churn patterns. Customers with shorter tenure, higher monthly charges, Fiber Optic internet service, and month-to-month contracts exhibited higher churn rates. Correlation analysis confirmed a strong relationship between tenure and total charges. These visualizations support the predictive modeling results and provide valuable guidance for customer retention strategies.
